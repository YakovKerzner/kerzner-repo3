```python
# Ğ‘Ğ»Ğ¾Ğº 2.3.4: Ğ¢Ñ€Ğ¸ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ° Ñ€Ğ¸ÑÑƒÑÑ‚ ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ñ‹ Ğ¾Ğ´Ğ½Ğ¾Ğ²Ñ€ĞµĞ¼ĞµĞ½Ğ½Ğ¾

class Robot:
    """ĞŸÑ€Ğ¾Ğ´Ğ²Ğ¸Ğ½ÑƒÑ‚Ñ‹Ğ¹ Ñ€Ğ¾Ğ±Ğ¾Ñ‚ Ñ ĞºÑ€Ğ°ÑĞ¸Ğ²Ñ‹Ğ¼ Ğ²Ñ‹Ğ²Ğ¾Ğ´Ğ¾Ğ¼ Ğ¸Ğ½Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ†Ğ¸Ğ¸"""
    
    def __init__(self, name, start_x=0, start_y=0):
        """Ğ¡Ğ¾Ğ·Ğ´Ğ°ĞµĞ¼ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ° Ñ Ğ¸Ğ¼ĞµĞ½ĞµĞ¼ Ğ¸ ÑÑ‚Ğ°Ñ€Ñ‚Ğ¾Ğ²Ğ¾Ğ¹ Ğ¿Ğ¾Ğ·Ğ¸Ñ†Ğ¸ĞµĞ¹"""
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x  # Ğ·Ğ°Ğ¿Ğ¾Ğ¼Ğ¸Ğ½Ğ°ĞµĞ¼ Ğ´Ğ¾Ğ¼
        self.start_y = start_y
        self.steps_made = 0     # ÑÑ‡ĞµÑ‚Ñ‡Ğ¸Ğº ÑˆĞ°Ğ³Ğ¾Ğ²
        print(f"ğŸ¤– {name} ÑĞ¾Ğ·Ğ´Ğ°Ğ½ Ğ² Ğ¿Ğ¾Ğ·Ğ¸Ñ†Ğ¸Ğ¸ ({start_x}, {start_y})")
    
    def show_info(self):
        """ĞšÑ€Ğ°ÑĞ¸Ğ²Ñ‹Ğ¹ Ğ²Ñ‹Ğ²Ğ¾Ğ´ Ğ¸Ğ½Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ†Ğ¸Ğ¸ Ğ¾ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğµ"""
        print(f"â”Œ{'â”€' * 40}â”")
        print(f"â”‚ ğŸ¤– Ğ Ğ¾Ğ±Ğ¾Ñ‚: {self.name:<30} â”‚")
        print(f"â”‚ ğŸ“ ĞŸĞ¾Ğ·Ğ¸Ñ†Ğ¸Ñ: ({self.x}, {self.y}){' ' * (22 - len(str(self.x)) - len(str(self.y)))}â”‚")
        print(f"â”‚ ğŸ  Ğ”Ğ¾Ğ¼: ({self.start_x}, {self.start_y}){' ' * (26 - len(str(self.start_x)) - len(str(self.start_y)))}â”‚")
        print(f"â”‚ ğŸ‘£ Ğ¨Ğ°Ğ³Ğ¾Ğ² ÑĞ´ĞµĞ»Ğ°Ğ½Ğ¾: {self.steps_made:<18} â”‚")
        print(f"â””{'â”€' * 40}â”˜")
    
    def move_right(self, steps=1):
        """Ğ”Ğ²Ğ¸Ğ¶ĞµĞ½Ğ¸Ğµ Ğ²Ğ¿Ñ€Ğ°Ğ²Ğ¾ Ğ½Ğ° ÑƒĞºĞ°Ğ·Ğ°Ğ½Ğ½Ğ¾Ğµ ĞºĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ ÑˆĞ°Ğ³Ğ¾Ğ²"""
        if steps <= 0:
            print(f"âŒ {self.name}: Ğ½ĞµĞ»ÑŒĞ·Ñ Ğ´Ğ²Ğ¸Ğ³Ğ°Ñ‚ÑŒÑÑ Ğ½Ğ° {steps} ÑˆĞ°Ğ³Ğ¾Ğ²!")
            return
        
        for step in range(steps):
            self.x += 1
            self.steps_made += 1
            print(f"    â¡ï¸ {self.name} ÑˆĞ°Ğ³ {step+1} â†’ ({self.x}, {self.y})")
    
    def move_left(self, steps=1):
        """Ğ”Ğ²Ğ¸Ğ¶ĞµĞ½Ğ¸Ğµ Ğ²Ğ»ĞµĞ²Ğ¾ Ğ½Ğ° ÑƒĞºĞ°Ğ·Ğ°Ğ½Ğ½Ğ¾Ğµ ĞºĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ ÑˆĞ°Ğ³Ğ¾Ğ²"""
        if steps <= 0:
            print(f"âŒ {self.name}: Ğ½ĞµĞ»ÑŒĞ·Ñ Ğ´Ğ²Ğ¸Ğ³Ğ°Ñ‚ÑŒÑÑ Ğ½Ğ° {steps} ÑˆĞ°Ğ³Ğ¾Ğ²!")
            return
        
        for step in range(steps):
            self.x -= 1
            self.steps_made += 1
            print(f"    â¬…ï¸ {self.name} ÑˆĞ°Ğ³ {step+1} â†’ ({self.x}, {self.y})")
    
    def move_up(self, steps=1):
        """Ğ”Ğ²Ğ¸Ğ¶ĞµĞ½Ğ¸Ğµ Ğ²Ğ²ĞµÑ€Ñ… Ğ½Ğ° ÑƒĞºĞ°Ğ·Ğ°Ğ½Ğ½Ğ¾Ğµ ĞºĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ ÑˆĞ°Ğ³Ğ¾Ğ²"""
        if steps <= 0:
            print(f"âŒ {self.name}: Ğ½ĞµĞ»ÑŒĞ·Ñ Ğ´Ğ²Ğ¸Ğ³Ğ°Ñ‚ÑŒÑÑ Ğ½Ğ° {steps} ÑˆĞ°Ğ³Ğ¾Ğ²!")
            return
        
        for step in range(steps):
            self.y += 1
            self.steps_made += 1
            print(f"    â¬†ï¸ {self.name} ÑˆĞ°Ğ³ {step+1} â†’ ({self.x}, {self.y})")
    
    def move_down(self, steps=1):
        """Ğ”Ğ²Ğ¸Ğ¶ĞµĞ½Ğ¸Ğµ Ğ²Ğ½Ğ¸Ğ· Ğ½Ğ° ÑƒĞºĞ°Ğ·Ğ°Ğ½Ğ½Ğ¾Ğµ ĞºĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ ÑˆĞ°Ğ³Ğ¾Ğ²"""
        if steps <= 0:
            print(f"âŒ {self.name}: Ğ½ĞµĞ»ÑŒĞ·Ñ Ğ´Ğ²Ğ¸Ğ³Ğ°Ñ‚ÑŒÑÑ Ğ½Ğ° {steps} ÑˆĞ°Ğ³Ğ¾Ğ²!")
            return
        
        for step in range(steps):
            self.y -= 1
            self.steps_made += 1
            print(f"    â¬‡ï¸ {self.name} ÑˆĞ°Ğ³ {step+1} â†’ ({self.x}, {self.y})")
    
    def walk_square(self, size):
        """Ğ Ğ¾Ğ±Ğ¾Ñ‚ Ñ€Ğ¸ÑÑƒĞµÑ‚ ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚ Ğ·Ğ°Ğ´Ğ°Ğ½Ğ½Ğ¾Ğ³Ğ¾ Ñ€Ğ°Ğ·Ğ¼ĞµÑ€Ğ°"""
        print(f"\nğŸ¨ {self.name} Ğ½Ğ°Ñ‡Ğ¸Ğ½Ğ°ĞµÑ‚ Ñ€Ğ¸ÑĞ¾Ğ²Ğ°Ñ‚ÑŒ ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚ {size}x{size}")
        
        # Ğ’Ğ¿Ñ€Ğ°Ğ²Ğ¾
        print(f"  â¡ï¸ Ğ˜Ğ´ĞµĞ¼ Ğ²Ğ¿Ñ€Ğ°Ğ²Ğ¾ ({size} ÑˆĞ°Ğ³Ğ¾Ğ²):")
        self.move_right(size)
        
        # Ğ’Ğ²ĞµÑ€Ñ…
        print(f"  â¬†ï¸ Ğ˜Ğ´ĞµĞ¼ Ğ²Ğ²ĞµÑ€Ñ… ({size} ÑˆĞ°Ğ³Ğ¾Ğ²):")
        self.move_up(size)
        
        # Ğ’Ğ»ĞµĞ²Ğ¾
        print(f"  â¬…ï¸ Ğ˜Ğ´ĞµĞ¼ Ğ²Ğ»ĞµĞ²Ğ¾ ({size} ÑˆĞ°Ğ³Ğ¾Ğ²):")
        self.move_left(size)
        
        # Ğ’Ğ½Ğ¸Ğ·
        print(f"  â¬‡ï¸ Ğ˜Ğ´ĞµĞ¼ Ğ²Ğ½Ğ¸Ğ· ({size} ÑˆĞ°Ğ³Ğ¾Ğ²):")
        self.move_down(size)
        
        print(f"âœ… {self.name} Ğ·Ğ°Ğ²ĞµÑ€ÑˆĞ¸Ğ» ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚!")

# Ğ¡Ğ¾Ğ·Ğ´Ğ°ĞµĞ¼ Ñ‚Ñ€ĞµÑ… Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ¾Ğ² Ğ² Ñ€Ğ°Ğ·Ğ½Ñ‹Ñ… Ğ¿Ğ¾Ğ·Ğ¸Ñ†Ğ¸ÑÑ…
print("ğŸ­ Ğ—Ğ°Ğ¿ÑƒÑĞºĞ°ĞµĞ¼ Ñ„Ğ°Ğ±Ñ€Ğ¸ĞºÑƒ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ¾Ğ²!")

# TODO: Ğ¡Ğ¾Ğ·Ğ´Ğ°Ğ¹Ñ‚Ğµ Ñ‚Ñ€ĞµÑ… Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ¾Ğ²
# Ğ Ğ¾Ğ±Ğ¾Ñ‚ 1: "ĞĞ»ÑŒÑ„Ğ°" Ğ² Ğ¿Ğ¾Ğ·Ğ¸Ñ†Ğ¸Ğ¸ (0, 0)
robot1 = Robot("ĞĞ»ÑŒÑ„Ğ°", 0, 0)

# Ğ Ğ¾Ğ±Ğ¾Ñ‚ 2: "Ğ‘ĞµÑ‚Ğ°" Ğ² Ğ¿Ğ¾Ğ·Ğ¸Ñ†Ğ¸Ğ¸ (10, 5)  
robot2 = Robot("Ğ‘ĞµÑ‚Ğ°", 10, 5)

# Ğ Ğ¾Ğ±Ğ¾Ñ‚ 3: "Ğ“Ğ°Ğ¼Ğ¼Ğ°" Ğ² Ğ¿Ğ¾Ğ·Ğ¸Ñ†Ğ¸Ğ¸ (20, 10)
robot3 = Robot("Ğ“Ğ°Ğ¼Ğ¼Ğ°", 20, 10)

# ĞŸĞ¾ĞºĞ°Ğ·Ğ°Ñ‚ÑŒ Ğ½Ğ°Ñ‡Ğ°Ğ»ÑŒĞ½ÑƒÑ Ğ¸Ğ½Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ†Ğ¸Ñ
print("\nğŸ“Š ĞĞĞ§ĞĞ›Ğ¬ĞĞĞ• Ğ¡ĞĞ¡Ğ¢ĞĞ¯ĞĞ˜Ğ• Ğ ĞĞ‘ĞĞ¢ĞĞ’:")
# TODO: Ğ’Ñ‹Ğ·Ğ²Ğ°Ñ‚ÑŒ show_info() Ğ´Ğ»Ñ ĞºĞ°Ğ¶Ğ´Ğ¾Ğ³Ğ¾ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ°
robot1.show_info()
robot2.show_info()
robot3.show_info()

# Ğ—Ğ°Ğ¿ÑƒÑÑ‚Ğ¸Ñ‚ÑŒ Ğ²ÑĞµÑ… Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ¾Ğ² Ñ€Ğ¸ÑĞ¾Ğ²Ğ°Ñ‚ÑŒ
print("\nğŸ¨ Ğ’Ğ¡Ğ• Ğ ĞĞ‘ĞĞ¢Ğ« ĞĞĞ§Ğ˜ĞĞĞ®Ğ¢ Ğ Ğ˜Ğ¡ĞĞ’ĞĞ¢Ğ¬:")

# TODO: Ğ—Ğ°ÑÑ‚Ğ°Ğ²Ğ¸Ñ‚ÑŒ ĞºĞ°Ğ¶Ğ´Ğ¾Ğ³Ğ¾ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ° Ğ½Ğ°Ñ€Ğ¸ÑĞ¾Ğ²Ğ°Ñ‚ÑŒ ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚ Ñ€Ğ°Ğ·Ğ½Ğ¾Ğ³Ğ¾ Ñ€Ğ°Ğ·Ğ¼ĞµÑ€Ğ°
# ĞĞ»ÑŒÑ„Ğ° Ñ€Ğ¸ÑÑƒĞµÑ‚ ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚ 2x2
robot1.walk_square(2)

# Ğ‘ĞµÑ‚Ğ° Ñ€Ğ¸ÑÑƒĞµÑ‚ ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚ 3x3
robot2.walk_square(3)

# Ğ“Ğ°Ğ¼Ğ¼Ğ° Ñ€Ğ¸ÑÑƒĞµÑ‚ ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚ 4x4  
robot3.walk_square(4)

# ĞŸĞ¾ĞºĞ°Ğ·Ğ°Ñ‚ÑŒ Ñ„Ğ¸Ğ½Ğ°Ğ»ÑŒĞ½ÑƒÑ Ğ¸Ğ½Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ†Ğ¸Ñ
print("\nğŸ“Š Ğ¤Ğ˜ĞĞĞ›Ğ¬ĞĞĞ• Ğ¡ĞĞ¡Ğ¢ĞĞ¯ĞĞ˜Ğ• Ğ ĞĞ‘ĞĞ¢ĞĞ’:")
# TODO: Ğ¡Ğ½Ğ¾Ğ²Ğ° Ğ¿Ğ¾ĞºĞ°Ğ·Ğ°Ñ‚ÑŒ Ğ¸Ğ½Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ†Ğ¸Ñ Ğ¾ Ğ²ÑĞµÑ… Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ°Ñ…
robot1.show_info()
robot2.show_info()
robot3.show_info()

# ĞŸÑ€Ğ¾Ğ²ĞµÑ€ĞºĞ°: ĞºÑ‚Ğ¾ Ğ²ĞµÑ€Ğ½ÑƒĞ»ÑÑ Ğ´Ğ¾Ğ¼Ğ¾Ğ¹?
print("\nğŸ  ĞŸĞ ĞĞ’Ğ•Ğ ĞšĞ: ĞšĞ¢Ğ Ğ’Ğ•Ğ ĞĞ£Ğ›Ğ¡Ğ¯ Ğ”ĞĞœĞĞ™?")
robots = [robot1, robot2, robot3]
for robot in robots:
    if robot.x == robot.start_x and robot.y == robot.start_y:
        print(f"âœ… {robot.name} Ğ²ĞµÑ€Ğ½ÑƒĞ»ÑÑ Ğ´Ğ¾Ğ¼Ğ¾Ğ¹!")
    else:
        print(f"âŒ {robot.name} Ğ·Ğ°Ğ±Ğ»ÑƒĞ´Ğ¸Ğ»ÑÑ!")

print(f"\nğŸ‰ Ğ—Ğ°Ğ´Ğ°Ğ½Ğ¸Ğµ Ğ²Ñ‹Ğ¿Ğ¾Ğ»Ğ½ĞµĞ½Ğ¾! ĞœÑ‹ ÑƒĞ¿Ñ€Ğ°Ğ²Ğ»ÑĞ»Ğ¸ {len(robots)} Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ°Ğ¼Ğ¸ Ğ¾Ğ´Ğ½Ğ¾Ğ²Ñ€ĞµĞ¼ĞµĞ½Ğ½Ğ¾!")
```

```python
# Ğ‘Ğ»Ğ¾Ğº 3.1: Ğ¤Ğ°Ğ±Ñ€Ğ¸ĞºĞ° Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ¾Ğ² - Ğ¼Ğ°ÑÑĞ¾Ğ²Ğ¾Ğµ Ğ¿Ñ€Ğ¾Ğ¸Ğ·Ğ²Ğ¾Ğ´ÑÑ‚Ğ²Ğ¾!
# Ğ¡Ğ¾Ğ·Ğ´Ğ°ĞµĞ¼ Ğ¼Ğ½Ğ¾Ğ³Ğ¾ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ¾Ğ² Ğ¸Ğ· ÑĞ¿Ğ¸ÑĞºĞ° Ğ´Ğ°Ğ½Ğ½Ñ‹Ñ…

class Robot:
    """Ğ Ğ¾Ğ±Ğ¾Ñ‚ Ğ´Ğ»Ñ Ğ¼Ğ°ÑÑĞ¾Ğ²Ğ¾Ğ³Ğ¾ Ğ¿Ñ€Ğ¾Ğ¸Ğ·Ğ²Ğ¾Ğ´ÑÑ‚Ğ²Ğ°"""
    
    def __init__(self, name, start_x=0, start_y=0, color="blue"):
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x
        self.start_y = start_y
        self.color = color
        self.steps_made = 0
        print(f"ğŸ¤– {name} ({color}) ÑĞ¾Ğ·Ğ´Ğ°Ğ½ Ğ² ({start_x}, {start_y})")
    
    def show_info(self):
        """ĞšĞ¾Ğ¼Ğ¿Ğ°ĞºÑ‚Ğ½Ñ‹Ğ¹ Ğ²Ñ‹Ğ²Ğ¾Ğ´ Ğ¸Ğ½Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ†Ğ¸Ğ¸"""
        status = "ğŸ " if (self.x == self.start_x and self.y == self.start_y) else "ğŸ—ºï¸"
        print(f"{status} {self.name:<10} | ĞŸĞ¾Ğ·Ğ¸Ñ†Ğ¸Ñ: ({self.x:2}, {self.y:2}) | Ğ¨Ğ°Ğ³Ğ¾Ğ²: {self.steps_made:2} | Ğ¦Ğ²ĞµÑ‚: {self.color}")
    
    def walk_square(self, size=3):
        """Ğ‘Ñ‹ÑÑ‚Ñ€Ğ¾Ğµ Ñ€Ğ¸ÑĞ¾Ğ²Ğ°Ğ½Ğ¸Ğµ ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ğ° (Ğ±ĞµĞ· Ğ´ĞµÑ‚Ğ°Ğ»ÑŒĞ½Ğ¾Ğ³Ğ¾ Ğ²Ñ‹Ğ²Ğ¾Ğ´Ğ°)"""
        directions = ["right", "up", "left", "down"]
        
        for direction in directions:
            for step in range(size):
                if direction == "right":
                    self.x += 1
                elif direction == "up":
                    self.y += 1
                elif direction == "left":
                    self.x -= 1
                elif direction == "down":
                    self.y -= 1
                self.steps_made += 1
        
        print(f"âœ… {self.name} Ğ·Ğ°Ğ²ĞµÑ€ÑˆĞ¸Ğ» ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚ {size}x{size}")

# Ğ‘Ğ°Ğ·Ğ° Ğ´Ğ°Ğ½Ğ½Ñ‹Ñ… Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ¾Ğ² Ğ´Ğ»Ñ ÑĞ¾Ğ·Ğ´Ğ°Ğ½Ğ¸Ñ ĞºĞ¾Ğ½Ñ†ĞµĞ½Ñ‚Ñ€Ğ¸Ñ‡ĞµÑĞºĞ¸Ñ… ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ğ¾Ğ²
# ĞšĞ°Ğ¶Ğ´Ñ‹Ğ¹ ÑĞ»Ğ¾Ğ²Ğ°Ñ€ÑŒ Ğ´Ğ¾Ğ»Ğ¶ĞµĞ½ ÑĞ¾Ğ´ĞµÑ€Ğ¶Ğ°Ñ‚ÑŒ: "name", "x", "y", "color"
# Ğ¡Ñ‚Ğ°Ñ€Ñ‚Ğ¾Ğ²Ñ‹Ğµ Ğ¿Ğ¾Ğ·Ğ¸Ñ†Ğ¸Ğ¸ Ğ¸ Ñ€Ğ°Ğ·Ğ¼ĞµÑ€Ñ‹ Ğ¿Ğ¾Ğ´Ğ¾Ğ±Ñ€Ğ°Ğ½Ñ‹ Ñ‚Ğ°Ğº, Ñ‡Ñ‚Ğ¾Ğ±Ñ‹ Ğ¿Ğ¾Ğ»ÑƒÑ‡Ğ¸Ğ»Ğ¸ÑÑŒ ĞºĞ¾Ğ½Ñ†ĞµĞ½Ñ‚Ñ€Ğ¸Ñ‡ĞµÑĞºĞ¸Ğµ ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ñ‹
robots_data = [
    # TODO: Ğ—Ğ°Ğ¿Ğ¾Ğ»Ğ½Ğ¸Ñ‚Ğµ ÑĞ»Ğ¾Ğ²Ğ°Ñ€Ğ¸ Ğ´Ğ°Ğ½Ğ½Ñ‹Ğ¼Ğ¸ Ğ´Ğ»Ñ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ¾Ğ²
    # Ğ¡Ñ‚Ğ°Ñ€Ñ‚Ğ¾Ğ²Ñ‹Ğµ Ğ¿Ğ¾Ğ·Ğ¸Ñ†Ğ¸Ğ¸: (0,0), (1,1), (2,2)
    # Ğ¦Ğ²ĞµÑ‚Ğ° Ğ¼Ğ¾Ğ¶ĞµÑ‚Ğµ Ğ²Ñ‹Ğ±Ñ€Ğ°Ñ‚ÑŒ Ğ»ÑĞ±Ñ‹Ğµ
    {"name": "R1D1", "x": 0, "y": 0, "color": "blue" },
    {"name": "R2D2", "x": 1, "y": 1, "color": "red" },
    {"name": "R3D3", "x": 2, "y": 2, "color": "green" },
]

print("ğŸ­ Ğ¤ĞĞ‘Ğ Ğ˜ĞšĞ Ğ ĞĞ‘ĞĞ¢ĞĞ’ Ğ—ĞĞŸĞ£Ğ©Ğ•ĞĞ!")
print(f"ğŸ“‹ Ğ’ Ğ¾Ñ‡ĞµÑ€ĞµĞ´Ğ¸ Ğ½Ğ° Ğ¿Ñ€Ğ¾Ğ¸Ğ·Ğ²Ğ¾Ğ´ÑÑ‚Ğ²Ğ¾: {len(robots_data)} Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ¾Ğ²")

# TODO: Ğ¡Ğ¾Ğ·Ğ´Ğ°Ğ¹Ñ‚Ğµ Ğ²ÑĞµÑ… Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ¾Ğ² Ğ¸Ğ· ÑĞ¿Ğ¸ÑĞºĞ° Ğ´Ğ°Ğ½Ğ½Ñ‹Ñ…
robot_army = []  # ÑĞ¿Ğ¸ÑĞ¾Ğº Ğ´Ğ»Ñ Ñ…Ñ€Ğ°Ğ½ĞµĞ½Ğ¸Ñ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ¾Ğ²

print("\nğŸ”§ ĞŸĞ ĞĞ˜Ğ—Ğ’ĞĞ”Ğ¡Ğ¢Ğ’Ğ Ğ ĞĞ‘ĞĞ¢ĞĞ’:")
for robot_data in robots_data:
    # TODO: Ğ¡Ğ¾Ğ·Ğ´Ğ°Ğ¹Ñ‚Ğµ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ° Ğ¸Ğ· Ğ´Ğ°Ğ½Ğ½Ñ‹Ñ… ÑĞ»Ğ¾Ğ²Ğ°Ñ€Ñ
    # ĞŸĞ¾Ğ´ÑĞºĞ°Ğ·ĞºĞ°: robot_data["name"], robot_data["x"], Ğ¸ Ñ‚.Ğ´.
    robot = Robot(
        robot_data["name"],    # Ğ¸Ğ¼Ñ
        robot_data["x"],    # x ĞºĞ¾Ğ¾Ñ€Ğ´Ğ¸Ğ½Ğ°Ñ‚Ğ°  
        robot_data["y"],    # y ĞºĞ¾Ğ¾Ñ€Ğ´Ğ¸Ğ½Ğ°Ñ‚Ğ°
        robot_data["color"]     # Ñ†Ğ²ĞµÑ‚
    )
    robot_army.append(robot)

print(f"\nâœ… ĞŸÑ€Ğ¾Ğ¸Ğ·Ğ²ĞµĞ´ĞµĞ½Ğ¾ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ¾Ğ²: {len(robot_army)}")

# ĞŸĞ¾ĞºĞ°Ğ·Ğ°Ñ‚ÑŒ Ğ¸Ğ½Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ†Ğ¸Ñ Ğ¾ Ğ²ÑĞµÑ… Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ°Ñ…
print("\nğŸ“Š Ğ Ğ•Ğ•Ğ¡Ğ¢Ğ  Ğ ĞĞ‘ĞĞ¢ĞĞ’:")
print("   Ğ¡Ñ‚Ğ°Ñ‚ÑƒÑ | Ğ˜Ğ¼Ñ        | ĞŸĞ¾Ğ·Ğ¸Ñ†Ğ¸Ñ    | Ğ¨Ğ°Ğ³Ğ¾Ğ² | Ğ¦Ğ²ĞµÑ‚")
print("   " + "â”€" * 50)

# TODO: ĞŸĞ¾ĞºĞ°Ğ¶Ğ¸Ñ‚Ğµ Ğ¸Ğ½Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ†Ğ¸Ñ Ğ¾ ĞºĞ°Ğ¶Ğ´Ğ¾Ğ¼ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğµ
for robot in robot_army:
    robot.show_info()

# Ğ—Ğ°Ğ¿ÑƒÑÑ‚Ğ¸Ñ‚ÑŒ Ğ²ÑĞµÑ… Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ¾Ğ² Ñ€Ğ°Ğ±Ğ¾Ñ‚Ğ°Ñ‚ÑŒ
print("\nğŸ¨ ĞœĞĞ¡Ğ¡ĞĞ’ĞĞ• ĞŸĞ ĞĞ˜Ğ—Ğ’ĞĞ”Ğ¡Ğ¢Ğ’Ğ ĞšĞ’ĞĞ”Ğ ĞĞ¢ĞĞ’:")

# TODO: Ğ—Ğ°ÑÑ‚Ğ°Ğ²ÑŒÑ‚Ğµ Ğ²ÑĞµÑ… Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ¾Ğ² Ğ½Ğ°Ñ€Ğ¸ÑĞ¾Ğ²Ğ°Ñ‚ÑŒ ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ñ‹ Ñ€Ğ°Ğ·Ğ½Ğ¾Ğ³Ğ¾ Ñ€Ğ°Ğ·Ğ¼ĞµÑ€Ğ°
# Ğ”Ğ»Ñ ĞºĞ¾Ğ½Ñ†ĞµĞ½Ñ‚Ñ€Ğ¸Ñ‡ĞµÑĞºĞ¸Ñ… ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ğ¾Ğ² Ğ¸ÑĞ¿Ğ¾Ğ»ÑŒĞ·ÑƒĞ¹Ñ‚Ğµ Ñ€Ğ°Ğ·Ğ¼ĞµÑ€Ñ‹: 4, 3, 2
sizes = [4, 3, 2]  # Ñ€Ğ°Ğ·Ğ¼ĞµÑ€Ñ‹ ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ğ¾Ğ² Ğ´Ğ»Ñ ĞºĞ¾Ğ½Ñ†ĞµĞ½Ñ‚Ñ€Ğ¸Ñ‡ĞµÑĞºĞ¸Ñ… ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚Ğ¾Ğ²

for i, robot in enumerate(robot_army):
    size = sizes[i]
    print(f"ğŸ¤– {robot.name} Ñ€Ğ¸ÑÑƒĞµÑ‚ ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚ {size}x{size}...")
    robot.walk_square(size)

# Ğ¤Ğ¸Ğ½Ğ°Ğ»ÑŒĞ½Ğ°Ñ ÑÑ‚Ğ°Ñ‚Ğ¸ÑÑ‚Ğ¸ĞºĞ°
print("\nğŸ“Š Ğ¤Ğ˜ĞĞĞ›Ğ¬ĞĞĞ¯ Ğ¡Ğ¢ĞĞ¢Ğ˜Ğ¡Ğ¢Ğ˜ĞšĞ:")
print("   Ğ¡Ñ‚Ğ°Ñ‚ÑƒÑ | Ğ˜Ğ¼Ñ        | ĞŸĞ¾Ğ·Ğ¸Ñ†Ğ¸Ñ    | Ğ¨Ğ°Ğ³Ğ¾Ğ² | Ğ¦Ğ²ĞµÑ‚") 
print("   " + "â”€" * 50)

# TODO: Ğ¡Ğ½Ğ¾Ğ²Ğ° Ğ¿Ğ¾ĞºĞ°Ğ¶Ğ¸Ñ‚Ğµ Ğ¸Ğ½Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ†Ğ¸Ñ Ğ¾ Ğ²ÑĞµÑ… Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ°Ñ…


# ĞĞ½Ğ°Ğ»Ğ¸Ğ· Ñ€ĞµĞ·ÑƒĞ»ÑŒÑ‚Ğ°Ñ‚Ğ¾Ğ²  
print(f"\nğŸ“ˆ ĞĞĞĞ›Ğ˜Ğ— ĞŸĞ ĞĞ˜Ğ—Ğ’ĞĞ”Ğ¡Ğ¢Ğ’Ğ:")

# TODO: ĞŸĞ¾Ğ´ÑÑ‡Ğ¸Ñ‚Ğ°Ğ¹Ñ‚Ğµ ÑÑ‚Ğ°Ñ‚Ğ¸ÑÑ‚Ğ¸ĞºÑƒ
total_steps = 0
robots_at_home = 0

for robot in robot_army:
    total_steps += robot.steps_made  # Ğ´Ğ¾Ğ±Ğ°Ğ²ÑŒÑ‚Ğµ ÑˆĞ°Ğ³Ğ¸ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ°
    
    # ĞŸÑ€Ğ¾Ğ²ĞµÑ€ÑŒÑ‚Ğµ, Ğ´Ğ¾Ğ¼Ğ° Ğ»Ğ¸ Ñ€Ğ¾Ğ±Ğ¾Ñ‚
    if robot.x == robot.start_x and robot.y == robot.start_y:
        robots_at_home += 1

print(f"   ğŸ¤– Ğ’ÑĞµĞ³Ğ¾ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ¾Ğ²: {len(robot_army)}")
print(f"   ğŸ‘£ ĞĞ±Ñ‰ĞµĞµ ĞºĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ ÑˆĞ°Ğ³Ğ¾Ğ²: {total_steps}")
print(f"   ğŸ“Š Ğ¡Ñ€ĞµĞ´Ğ½ĞµĞµ ÑˆĞ°Ğ³Ğ¾Ğ² Ğ½Ğ° Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ°: {total_steps / len(robot_army):.1f}")
print(f"   ğŸ  Ğ Ğ¾Ğ±Ğ¾Ñ‚Ğ¾Ğ² Ğ²ĞµÑ€Ğ½ÑƒĞ»Ğ¾ÑÑŒ Ğ´Ğ¾Ğ¼Ğ¾Ğ¹: {robots_at_home}")
print(f"   ğŸ“ Ğ­Ñ„Ñ„ĞµĞºÑ‚Ğ¸Ğ²Ğ½Ğ¾ÑÑ‚ÑŒ Ğ²Ğ¾Ğ·Ğ²Ñ€Ğ°Ñ‚Ğ°: {robots_at_home / len(robot_army) * 100:.1f}%")

# ĞŸĞ¾Ğ¸ÑĞº Ñ‡ĞµĞ¼Ğ¿Ğ¸Ğ¾Ğ½Ğ¾Ğ²
most_active = max(robot_army, key=lambda r: r.steps_made)
print(f"   ğŸ† Ğ¡Ğ°Ğ¼Ñ‹Ğ¹ Ğ°ĞºÑ‚Ğ¸Ğ²Ğ½Ñ‹Ğ¹: {most_active.name} ({most_active.steps_made} ÑˆĞ°Ğ³Ğ¾Ğ²)")

print(f"\nğŸ‰ Ğ¤ĞĞ‘Ğ Ğ˜ĞšĞ Ğ—ĞĞ’Ğ•Ğ Ğ¨Ğ˜Ğ›Ğ Ğ ĞĞ‘ĞĞ¢Ğ£! Ğ’ÑĞµ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ñ‹ Ğ³Ğ¾Ñ‚Ğ¾Ğ²Ñ‹ Ğº ÑĞ»ÑƒĞ¶Ğ±Ğµ!")
```

```python
# Ğ‘Ğ»Ğ¾Ğº 3.2.1: Ğ Ğ¸ÑĞ¾Ğ²Ğ°Ğ½Ğ¸Ğµ Ğ¾Ğ´Ğ½Ğ¾Ğ³Ğ¾ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ° Ñ Ğ¿Ğ¾Ğ¼Ğ¾Ñ‰ÑŒÑ matplotlib
# Ğ£Ñ‡Ğ¸Ğ¼ÑÑ Ñ€Ğ¸ÑĞ¾Ğ²Ğ°Ñ‚ÑŒ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ° Ğ½Ğ° ÑĞºÑ€Ğ°Ğ½Ğµ!

import matplotlib.pyplot as plt
import numpy as np

class Robot:
    """Ğ Ğ¾Ğ±Ğ¾Ñ‚, ĞºĞ¾Ñ‚Ğ¾Ñ€Ñ‹Ğ¹ ÑƒĞ¼ĞµĞµÑ‚ Ñ€Ğ¸ÑĞ¾Ğ²Ğ°Ñ‚ÑŒ ÑĞµĞ±Ñ Ğ½Ğ° ÑĞºÑ€Ğ°Ğ½Ğµ"""
    
    def __init__(self, name, start_x=0, start_y=0, color="blue"):
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x
        self.start_y = start_y
        self.color = color
        self.steps_made = 0
        print(f"ğŸ¤– {name} ({color}) ÑĞ¾Ğ·Ğ´Ğ°Ğ½ Ğ² ({start_x}, {start_y})")
    
    def draw(self, size=100):
        """Ğ Ğ¸ÑÑƒĞµÑ‚ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ° Ğ½Ğ° Ñ‚ĞµĞºÑƒÑ‰ĞµĞ¼ Ğ³Ñ€Ğ°Ñ„Ğ¸ĞºĞµ"""
        # TODO: ĞĞ°Ñ€Ğ¸ÑÑƒĞ¹Ñ‚Ğµ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ° ĞºĞ°Ğº Ñ‚Ğ¾Ñ‡ĞºÑƒ
        # ĞŸĞ¾Ğ´ÑĞºĞ°Ğ·ĞºĞ°: Ğ¸ÑĞ¿Ğ¾Ğ»ÑŒĞ·ÑƒĞ¹Ñ‚Ğµ plt.scatter()
        plt.scatter(self.x, self.y, 
                   color=self.color, 
                   s=size, 
                   alpha=0.8,
                   label=self.name)
        
        # TODO: Ğ”Ğ¾Ğ±Ğ°Ğ²ÑŒÑ‚Ğµ Ğ¿Ğ¾Ğ´Ğ¿Ğ¸ÑÑŒ Ñ Ğ¸Ğ¼ĞµĞ½ĞµĞ¼ Ğ¸ ĞºĞ¾Ğ¾Ñ€Ğ´Ğ¸Ğ½Ğ°Ñ‚Ğ°Ğ¼Ğ¸
        # ĞŸĞ¾Ğ´ÑĞºĞ°Ğ·ĞºĞ°: Ğ¸ÑĞ¿Ğ¾Ğ»ÑŒĞ·ÑƒĞ¹Ñ‚Ğµ plt.annotate()
        plt.annotate(f'{self.name}\n({self.x},{self.y})', 
                    (self.x, self.y), 
                    xytext=(10, 10), 
                    textcoords='offset points',
                    fontsize=8,
                    bbox=dict(boxstyle="round,pad=0.3", 
                             facecolor=self.color, 
                             alpha=0.3))
    
    def move_right(self, steps=1):
        """Ğ”Ğ²Ğ¸Ğ¶ĞµĞ½Ğ¸Ğµ Ğ²Ğ¿Ñ€Ğ°Ğ²Ğ¾ Ğ½Ğ° ÑƒĞºĞ°Ğ·Ğ°Ğ½Ğ½Ğ¾Ğµ ĞºĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ ÑˆĞ°Ğ³Ğ¾Ğ²"""
        if steps <= 0:
            print(f"âŒ {self.name}: Ğ½ĞµĞ»ÑŒĞ·Ñ Ğ´Ğ²Ğ¸Ğ³Ğ°Ñ‚ÑŒÑÑ Ğ½Ğ° {steps} ÑˆĞ°Ğ³Ğ¾Ğ²!")
            return
        
        for step in range(steps):
            self.x += 1
            self.steps_made += 1
            print(f"    â¡ï¸ {self.name} ÑˆĞ°Ğ³ {step+1} â†’ ({self.x}, {self.y})")
    
    def move_left(self, steps=1):
        """Ğ”Ğ²Ğ¸Ğ¶ĞµĞ½Ğ¸Ğµ Ğ²Ğ»ĞµĞ²Ğ¾ Ğ½Ğ° ÑƒĞºĞ°Ğ·Ğ°Ğ½Ğ½Ğ¾Ğµ ĞºĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ ÑˆĞ°Ğ³Ğ¾Ğ²"""
        if steps <= 0:
            print(f"âŒ {self.name}: Ğ½ĞµĞ»ÑŒĞ·Ñ Ğ´Ğ²Ğ¸Ğ³Ğ°Ñ‚ÑŒÑÑ Ğ½Ğ° {steps} ÑˆĞ°Ğ³Ğ¾Ğ²!")
            return
        
        for step in range(steps):
            self.x -= 1
            self.steps_made += 1
            print(f"    â¬…ï¸ {self.name} ÑˆĞ°Ğ³ {step+1} â†’ ({self.x}, {self.y})")
    
    def move_up(self, steps=1):
        """Ğ”Ğ²Ğ¸Ğ¶ĞµĞ½Ğ¸Ğµ Ğ²Ğ²ĞµÑ€Ñ… Ğ½Ğ° ÑƒĞºĞ°Ğ·Ğ°Ğ½Ğ½Ğ¾Ğµ ĞºĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ ÑˆĞ°Ğ³Ğ¾Ğ²"""
        if steps <= 0:
            print(f"âŒ {self.name}: Ğ½ĞµĞ»ÑŒĞ·Ñ Ğ´Ğ²Ğ¸Ğ³Ğ°Ñ‚ÑŒÑÑ Ğ½Ğ° {steps} ÑˆĞ°Ğ³Ğ¾Ğ²!")
            return
        
        for step in range(steps):
            self.y += 1
            self.steps_made += 1
            print(f"    â¬†ï¸ {self.name} ÑˆĞ°Ğ³ {step+1} â†’ ({self.x}, {self.y})")
    
    def move_down(self, steps=1):
        """Ğ”Ğ²Ğ¸Ğ¶ĞµĞ½Ğ¸Ğµ Ğ²Ğ½Ğ¸Ğ· Ğ½Ğ° ÑƒĞºĞ°Ğ·Ğ°Ğ½Ğ½Ğ¾Ğµ ĞºĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ ÑˆĞ°Ğ³Ğ¾Ğ²"""
        if steps <= 0:
            print(f"âŒ {self.name}: Ğ½ĞµĞ»ÑŒĞ·Ñ Ğ´Ğ²Ğ¸Ğ³Ğ°Ñ‚ÑŒÑÑ Ğ½Ğ° {steps} ÑˆĞ°Ğ³Ğ¾Ğ²!")
            return
        
        for step in range(steps):
            self.y -= 1
            self.steps_made += 1
            print(f"    â¬‡ï¸ {self.name} ÑˆĞ°Ğ³ {step+1} â†’ ({self.x}, {self.y})")
    
    def walk_square(self, size=3):
        """Ğ Ğ¾Ğ±Ğ¾Ñ‚ Ñ€Ğ¸ÑÑƒĞµÑ‚ ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚ Ğ·Ğ°Ğ´Ğ°Ğ½Ğ½Ğ¾Ğ³Ğ¾ Ñ€Ğ°Ğ·Ğ¼ĞµÑ€Ğ°"""
        print(f"\nğŸ¨ {self.name} Ğ½Ğ°Ñ‡Ğ¸Ğ½Ğ°ĞµÑ‚ Ñ€Ğ¸ÑĞ¾Ğ²Ğ°Ñ‚ÑŒ ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚ {size}x{size}")
        
        # Ğ’Ğ¿Ñ€Ğ°Ğ²Ğ¾
        print(f"  â¡ï¸ Ğ˜Ğ´ĞµĞ¼ Ğ²Ğ¿Ñ€Ğ°Ğ²Ğ¾ ({size} ÑˆĞ°Ğ³Ğ¾Ğ²):")
        self.move_right(size)
        
        # Ğ’Ğ²ĞµÑ€Ñ…
        print(f"  â¬†ï¸ Ğ˜Ğ´ĞµĞ¼ Ğ²Ğ²ĞµÑ€Ñ… ({size} ÑˆĞ°Ğ³Ğ¾Ğ²):")
        self.move_up(size)
        
        # Ğ’Ğ»ĞµĞ²Ğ¾
        print(f"  â¬…ï¸ Ğ˜Ğ´ĞµĞ¼ Ğ²Ğ»ĞµĞ²Ğ¾ ({size} ÑˆĞ°Ğ³Ğ¾Ğ²):")
        self.move_left(size)
        
        # Ğ’Ğ½Ğ¸Ğ·
        print(f"  â¬‡ï¸ Ğ˜Ğ´ĞµĞ¼ Ğ²Ğ½Ğ¸Ğ· ({size} ÑˆĞ°Ğ³Ğ¾Ğ²):")
        self.move_down(size)
        
        print(f"âœ… {self.name} Ğ·Ğ°Ğ²ĞµÑ€ÑˆĞ¸Ğ» ĞºĞ²Ğ°Ğ´Ñ€Ğ°Ñ‚!")

# Ğ¢ĞµÑÑ‚: Ğ Ğ¸ÑÑƒĞµĞ¼ Ğ¾Ğ´Ğ½Ğ¾Ğ³Ğ¾ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ°
print("ğŸ¨ Ğ¢Ğ•Ğ¡Ğ¢: Ğ Ğ¸ÑĞ¾Ğ²Ğ°Ğ½Ğ¸Ğµ Ğ¾Ğ´Ğ½Ğ¾Ğ³Ğ¾ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ°")

# TODO: Ğ¡Ğ¾Ğ·Ğ´Ğ°Ğ¹Ñ‚Ğµ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ° Ğ´Ğ»Ñ Ñ‚ĞµÑÑ‚Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ¸Ñ
test_robot = Robot("R2D2", 0, 0, "red")

# ĞĞ°ÑÑ‚Ñ€Ğ°Ğ¸Ğ²Ğ°ĞµĞ¼ Ğ³Ñ€Ğ°Ñ„Ğ¸Ğº
plt.figure(figsize=(8, 6))

# TODO: ĞĞ°Ñ€Ğ¸ÑÑƒĞ¹Ñ‚Ğµ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ°
test_robot.draw()

# ĞĞ°ÑÑ‚Ñ€Ğ¾Ğ¹ĞºĞ¸ Ğ³Ñ€Ğ°Ñ„Ğ¸ĞºĞ°
plt.grid(True, alpha=0.3)
plt.xlim(-2, 10)
plt.ylim(-2, 8)
plt.title("ğŸ¤– ĞœĞ¾Ğ¹ Ğ¿ĞµÑ€Ğ²Ñ‹Ğ¹ Ñ€Ğ¾Ğ±Ğ¾Ñ‚ Ğ½Ğ° ÑĞºÑ€Ğ°Ğ½Ğµ!")
plt.xlabel("X ĞºĞ¾Ğ¾Ñ€Ğ´Ğ¸Ğ½Ğ°Ñ‚Ğ°")
plt.ylabel("Y ĞºĞ¾Ğ¾Ñ€Ğ´Ğ¸Ğ½Ğ°Ñ‚Ğ°")
plt.legend()
plt.show()

print("âœ… Ğ¢ĞµÑÑ‚ Ğ·Ğ°Ğ²ĞµÑ€ÑˆĞµĞ½! Ğ Ğ¾Ğ±Ğ¾Ñ‚ Ğ½Ğ°Ñ€Ğ¸ÑĞ¾Ğ²Ğ°Ğ½!")

# Ğ”Ğ¾Ğ¿Ğ¾Ğ»Ğ½Ğ¸Ñ‚ĞµĞ»ÑŒĞ½Ğ¾Ğµ Ğ·Ğ°Ğ´Ğ°Ğ½Ğ¸Ğµ: Ğ¿Ğ¾Ğ¿Ñ€Ğ¾Ğ±ÑƒĞ¹Ñ‚Ğµ Ñ€Ğ°Ğ·Ğ½Ñ‹Ğµ Ñ†Ğ²ĞµÑ‚Ğ° Ğ¸ Ñ€Ğ°Ğ·Ğ¼ĞµÑ€Ñ‹
print("\nğŸ¨ Ğ”ĞĞŸĞĞ›ĞĞ˜Ğ¢Ğ•Ğ›Ğ¬ĞĞĞ• Ğ—ĞĞ”ĞĞĞ˜Ğ•:")
print("Ğ¡Ğ¾Ğ·Ğ´Ğ°Ğ¹Ñ‚Ğµ ĞµÑ‰Ğµ Ğ¾Ğ´Ğ½Ğ¾Ğ³Ğ¾ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ° Ñ Ğ´Ñ€ÑƒĞ³Ğ¸Ğ¼ Ñ†Ğ²ĞµÑ‚Ğ¾Ğ¼ Ğ¸ Ñ€Ğ°Ğ·Ğ¼ĞµÑ€Ğ¾Ğ¼!")

# TODO: Ğ¡Ğ¾Ğ·Ğ´Ğ°Ğ¹Ñ‚Ğµ Ğ²Ñ‚Ğ¾Ñ€Ğ¾Ğ³Ğ¾ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ°
second_robot = Robot("R3D3", 6, 6, "green")

plt.figure(figsize=(8, 6))
# TODO: ĞĞ°Ñ€Ğ¸ÑÑƒĞ¹Ñ‚Ğµ Ğ²Ñ‚Ğ¾Ñ€Ğ¾Ğ³Ğ¾ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ° Ñ Ğ±Ğ¾Ğ»ÑŒÑˆĞ¸Ğ¼ Ñ€Ğ°Ğ·Ğ¼ĞµÑ€Ğ¾Ğ¼
second_robot.draw(size=150)

plt.grid(True, alpha=0.3)
plt.xlim(-2, 10)
plt.ylim(-2, 8)
plt.title("ğŸ¤–ğŸ¤– Ğ”Ğ²Ğ° Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ° Ğ½Ğ° ÑĞºÑ€Ğ°Ğ½Ğµ!")
plt.xlabel("X ĞºĞ¾Ğ¾Ñ€Ğ´Ğ¸Ğ½Ğ°Ñ‚Ğ°")
plt.ylabel("Y ĞºĞ¾Ğ¾Ñ€Ğ´Ğ¸Ğ½Ğ°Ñ‚Ğ°")
plt.legend()
plt.show()

print("ğŸ‰ ĞÑ‚Ğ»Ğ¸Ñ‡Ğ½Ğ¾! Ğ¢ĞµĞ¿ĞµÑ€ÑŒ Ğ²Ñ‹ ÑƒĞ¼ĞµĞµÑ‚Ğµ Ñ€Ğ¸ÑĞ¾Ğ²Ğ°Ñ‚ÑŒ Ñ€Ğ¾Ğ±Ğ¾Ñ‚Ğ¾Ğ²!")
```
